```python
# Importing libraries for calculations, data handling and visualization
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ENVIRONMENT SETUP

# Defining queue limits (maximum # of customers)
MAX_Q1 = 5  # Important customers
MAX_Q2 = 5  # Normal customers

# # of service agents
NUM_AGENTS = 2

# Defining possible actions
ACTIONS = {
    0: "Assign q1 to Agent 1",  # Assigning important customer to Agent 1
    1: "Assign q2 to Agent 2",  # Assigning normal customer to Agent 2
    2: "Assign q1 to Agent 2",  # Assigning important customer to Agent 2
    3: "Do nothing"             # No action taken
}
NUM_ACTIONS = len(ACTIONS)

# Probability of customer arrival in each queue
ARRIVAL_PROB = {
    "important": 0.3,
    "normal": 0.5
}

# Reward and penalty constants
REWARD_SERVE_IMPORTANT = 2.0
REWARD_SERVE_NORMAL = 1.0
PENALTY_MISMATCH = -2.0             # Wrong agent serving customer
PENALTY_IDLE_AGENT = -2.0           # Agents not serving when customers are waiting
PENALTY_WAITING_CUSTOMER = -1.0     # Customers still in queue
INVALID_ACTION_PENALTY = -10.0      # Trying to serve with busy agent or no customer

# Discount factor for future rewards
GAMMA = 0.9

# REWARD FUNCTION
def get_reward(state, action):

    #Calculates reward for a given state action pair.

    q1, q2, b1, b2 = state  # current state
    reward = 0

    # Check which action is taken and compute reward accordingly
    if action == 0:  # Assign q1 to Agent 1
        if q1 > 0 and b1 == 0:
            reward += REWARD_SERVE_IMPORTANT + PENALTY_MISMATCH
        elif q1 > 0 and b1 == 1 or q1 == 0:
            reward = INVALID_ACTION_PENALTY

    elif action == 1:  # Assign q2 to Agent 2
        if q2 > 0 and b2 == 0:
            reward += REWARD_SERVE_NORMAL
        elif q2 > 0 and b2 == 1 or q2 == 0:
            reward = INVALID_ACTION_PENALTY

    elif action == 2:  # Assign q1 to Agent 2
        if q1 > 0 and b2 == 0:
            reward += REWARD_SERVE_IMPORTANT
        elif q1 > 0 and b2 == 1 or q1 == 0:
            reward = INVALID_ACTION_PENALTY

    elif action == 3:  # Do nothing
        pass

    # Apply penalties for waiting customers and idle agents
    reward += (q1 + q2) * PENALTY_WAITING_CUSTOMER
    reward += ((1 - b1) + (1 - b2)) * PENALTY_IDLE_AGENT

    return reward

# TRANSITION FUNCTION
def get_next_state_and_prob(state, action):

    #Simulates the environment and returns possible next states and their probabilities based on the action taken.

    q1, q2, b1, b2 = state

    # Assume current state becomes next unless changed
    next_q1 = q1
    next_q2 = q2
    next_b1 = b1
    next_b2 = b2

    # Simulating customer being served based on the action
    if action == 0 and q1 > 0 and b1 == 0:
        next_q1 = max(0, q1 - 1)
        next_b1 = 1
    elif action == 1 and q2 > 0 and b2 == 0:
        next_q2 = max(0, q2 - 1)
        next_b2 = 1
    elif action == 2 and q1 > 0 and b2 == 0:
        next_q1 = max(0, q1 - 1)
        next_b2 = 1

    # Dictionary to store probabilities of next states
    next_states_probs = {}

    # Looping over all combinations of new arrivals (0 or 1)
    for aq1 in [0, 1]:
        for aq2 in [0, 1]:
            # Calculating arrival probability
            prob = (ARRIVAL_PROB["important"] if aq1 == 1 else (1 - ARRIVAL_PROB["important"])) * \
                   (ARRIVAL_PROB["normal"] if aq2 == 1 else (1 - ARRIVAL_PROB["normal"]))

            # Next state after serving and new arrivals
            next_state = (
                min(MAX_Q1, next_q1 + aq1),
                min(MAX_Q2, next_q2 + aq2),
                next_b1 if action != 0 or q1 == 0 or b1 == 1 else 0,
                next_b2 if action not in [1, 2] or b2 == 1 or (action == 1 and q2 == 0) or (action == 2 and q1 == 0) else 0
            )

            # Storing in dictionary
            next_states_probs[next_state] = next_states_probs.get(next_state, 0) + prob

    return next_states_probs


# VALUE ITERATION ALGORITHM
def value_iteration(epsilon=0.01, max_iterations=100):

    #Using Value Iteration to compute optimal value function and policy.

    V = np.zeros((MAX_Q1 + 1, MAX_Q2 + 1, 2, 2))  # value function
    policy = np.zeros_like(V, dtype=int)         # optimal action for each state
    iteration = 0

    # Repeat until convergence
    while True:
        V_old = V.copy()
        delta = 0

        # Looping over all states
        for q1 in range(MAX_Q1 + 1):
            for q2 in range(MAX_Q2 + 1):
                for b1 in range(2):
                    for b2 in range(2):
                        state = (q1, q2, b1, b2)
                        max_q_value = -float('inf')
                        best_action = -1

                        # Try each possible action
                        for action in range(NUM_ACTIONS):
                            reward = get_reward(state, action)
                            expected_value = 0

                            # Get expected value of next state
                            for next_state, prob in get_next_state_and_prob(state, action).items():
                                nq1, nq2, nb1, nb2 = next_state
                                expected_value += prob * V_old[nq1, nq2, nb1, nb2]

                            q_value = reward + GAMMA * expected_value

                            # Store the best action
                            if q_value > max_q_value:
                                max_q_value = q_value
                                best_action = action

                        # Update value and policy
                        V[q1, q2, b1, b2] = max_q_value
                        policy[q1, q2, b1, b2] = best_action
                        delta = max(delta, abs(V[q1, q2, b1, b2] - V_old[q1, q2, b1, b2]))

        iteration += 1
        # Stop if value function change is very small
        if delta < epsilon or iteration >= max_iterations:
            break

    return policy, V


# VISUALIZATION FUNCTIONS
def visualize_policy(policy):

    #Displaying policy as text in a grid for idle agents (b1 = b2 = 0)

    fig, axes = plt.subplots(MAX_Q2 + 1, MAX_Q1 + 1, figsize=(12, 10), sharex=True, sharey=True)
    plt.suptitle("Optimal Action when Both Agents are Idle (b1=0, b2=0)", fontsize=16)

    for q1 in range(MAX_Q1 + 1):
        for q2 in range(MAX_Q2 + 1):
            action = policy[q1, q2, 0, 0]
            axes[q2, q1].text(0.5, 0.5, ACTIONS[action], ha='center', va='center', fontsize=10,
                              bbox=dict(facecolor='lightgray', edgecolor='black', boxstyle='round,pad=0.5'))
            if q1 == 0:
                axes[q2, q1].set_ylabel(f"Q2 = {q2}")
            if q2 == MAX_Q2:
                axes[q2, q1].set_xlabel(f"Q1 = {q1}")
            axes[q2, q1].set_xticks([])
            axes[q2, q1].set_yticks([])

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def display_value_function(V):

    #Printing value table for all (q1, q2) states when agents are idle.

    value_table = np.zeros((MAX_Q1 + 1, MAX_Q2 + 1))
    for q1 in range(MAX_Q1 + 1):
        for q2 in range(MAX_Q2 + 1):
            value_table[q1, q2] = V[q1, q2, 0, 0]

    df = pd.DataFrame(value_table, index=[f"Q1={i}" for i in range(MAX_Q1 + 1)],
                      columns=[f"Q2={j}" for j in range(MAX_Q2 + 1)])
    print("\nOptimal Value Function (when both agents are idle):")
    print(df)

def display_policy_table(policy):

    #Showing a policy table with actions for each (q1, q2) when both agents are idle.

    policy_table = np.empty((MAX_Q1 + 1, MAX_Q2 + 1), dtype=object)
    for q1 in range(MAX_Q1 + 1):
        for q2 in range(MAX_Q2 + 1):
            policy_table[q1, q2] = ACTIONS[policy[q1, q2, 0, 0]]

    df = pd.DataFrame(policy_table, index=[f"Q1={i}" for i in range(MAX_Q1 + 1)],
                      columns=[f"Q2={j}" for j in range(MAX_Q2 + 1)])
    print("\nOptimal Policy Table (when both agents are idle):")
    print(df)

def display_policy_table_for_agent_state(policy, agent_state=(0, 0)):

    #Displaying the optimal policy for a specific agent busy/idle state.

    b1, b2 = agent_state
    policy_table = np.empty((MAX_Q1 + 1, MAX_Q2 + 1), dtype=object)

    for q1 in range(MAX_Q1 + 1):
        for q2 in range(MAX_Q2 + 1):
            action = policy[q1, q2, b1, b2]
            policy_table[q1, q2] = ACTIONS[action]

    df = pd.DataFrame(policy_table, index=[f"Q1={i}" for i in range(MAX_Q1 + 1)],
                      columns=[f"Q2={j}" for j in range(MAX_Q2 + 1)])
    print(f"\nOptimal Policy Table (Agent State: b1={b1}, b2={b2}):")
    print(df)
    return df

def visualize_value_function_heatmap(V, agent_state=(0, 0)):

    #Ploting heatmap of the value function for a given agent state.

    value_heatmap = V[:, :, agent_state[0], agent_state[1]]
    plt.figure(figsize=(8, 6))
    sns.heatmap(value_heatmap, annot=True, fmt=".1f", cmap="viridis",
                xticklabels=[f"Q2={i}" for i in range(MAX_Q2 + 1)],
                yticklabels=[f"Q1={i}" for i in range(MAX_Q1 + 1)])

    plt.title(f"Optimal Value Function (Agent State: {agent_state})")
    plt.xlabel("Number of Normal Customers (Q2)")
    plt.ylabel("Number of Important Customers (Q1)")
    plt.show()

# MAIN EXECUTION

# Running value iteration to get optimal policy and value function
optimal_policy, optimal_value_function = value_iteration()

# Showing visualizations and policy information
visualize_policy(optimal_policy)
display_value_function(optimal_value_function)
display_policy_table(optimal_policy)

# Showing policies for different agent busy states
display_policy_table_for_agent_state(optimal_policy, agent_state=(1, 0))  # Agent 1 busy
display_policy_table_for_agent_state(optimal_policy, agent_state=(0, 1))  # Agent 2 busy
display_policy_table_for_agent_state(optimal_policy, agent_state=(1, 1))  # Both agents busy

# Showing heatmaps of value function for different agent states
visualize_value_function_heatmap(optimal_value_function, agent_state=(1, 0))
visualize_value_function_heatmap(optimal_value_function, agent_state=(0, 1))
visualize_value_function_heatmap(optimal_value_function, agent_state=(1, 1))
```
